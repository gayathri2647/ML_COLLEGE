```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("datasets/car.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=["Sell Price"])
y = df['Sell Price']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
encoded_labels = ohe.fit_transform(df[['Car Model']]).toarray()
print(encoded_labels)
```

```python
dum = pd.DataFrame(encoded_labels)
dum.head()
```

```python
# dum = pd.get_dummies(x["Car Model"])
# dum.head()
```

```python
combained = pd.concat([x,dum], axis = 1)
combained.head()
```

```python
final_x = combained.drop(columns=["Car Model","Age"])
final_x
```

```python
print(final_x.columns)
```

```python
# Convert all column names to strings to avoid mixed type error
final_x.columns = final_x.columns.astype(str)
print(final_x.columns)
from sklearn.model_selection import train_test_split
```

```python
print(final_x.columns)  # Verify column names are strings
x_train,x_test,y_train,y_test = train_test_split(final_x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
print(f"r2 score is {r2_score(y_test,y_pred)}")
print(f"mean_absolute_error is {mean_absolute_error(y_test,y_pred)}")
```
