```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
d = df.select_dtypes(include=['object'])
d
```

```python
d.info()
```

```python
encode = ohe.fit_transform(d)
print(encode)
encoded_labels = ohe.fit_transform(d).toarray()
print(encoded_labels)
```

```python
df_encoded = pd.DataFrame(encoded_labels)
df_encoded
```

```python
df.drop(columns=d.columns,inplace=True)
df
```

```python
df1 = pd.concat([df_encoded,df],ignore_index=True, axis=1)
df1
```

```python
df1.shape
```

```python
X = df1.iloc[:,:-1]
y = df1.iloc[:,-1]
print(X)
print(y)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1, random_state=42)

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```
