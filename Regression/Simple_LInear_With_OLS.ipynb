# With Own Dataset




```python
import numpy as np
import statsmodels.api as s
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car_price_dataset_100_rows.csv")
```

```python
X = df[['Mileage (km)']]
y = df['Price ($1000s)']
```

```python
plt.scatter(X,y,c='black')
plt.xlabel('Mileage(km)')
plt.ylabel('Price')
plt.show()
```

```python
x = s.add_constant(X)
model = s.OLS(y,X).fit()
print(model.summary())
```
